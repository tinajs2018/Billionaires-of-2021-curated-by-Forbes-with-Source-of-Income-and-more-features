```python
import pandas as  pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
df=pd.read_csv("Billionaire.csv")
df.head(2)
```

```python
df_networth=df['NetWorth'].str.strip("$, B")

df_networth=df_networth.to_frame()
df_networth.head()
```

```python
df_networth.info()
```

```python
df_networth=pd.to_numeric(df_networth.NetWorth).astype(int)
df_networth
```

hence the above data is now into integer

```python
df_networth
```

```python
#droping Worththe Net
NetWorth=df.drop(['NetWorth'],axis=1)
```

```python
NetWorth.head(50)
```

```python
NetWorth.tail(50)
```

```python
NetWorth['df_networth']= df_networth
NetWorth.head(4)
```

```python
all_networt=pd.concat([df_networth,NetWorth],axis=1)
all_networt=all_networt[['Name','df_networth','Country','Source','Rank']]
all_networt
```

```python
#checking the data types of the data
print(all_networt.dtypes)
```

```python
#checking the highest worth by name
Name_top=all_networt[['Name','df_networth']].sort_values('df_networth',ascending=False)
Name_top.head(15)
```

```python
fig,ax=plt.subplots(figsize=(16,8))

ax.vlines(x=Name_top.index,ymin=0,ymax=Name_top.df_networth,color='Tomato',linewidth=5,alpha=0.5)

ax.scatter(x=Name_top.index, y=Name_top.df_networth,s=200,color='red')
ax.set_xticks(Name_top.index)
# ax.set_xticklabels(Name_top.Name.str.upper(),rotation=90,fontsize=15)
plt.show()
```

```python
#checking out newwrorth in regards to  countries
Net_sum=all_networt.groupby(['Country']).sum()
# highest_count=all_networt[['Country','df_networth']].sort_values('df_networth',ascending=False)
# highest_count

Net_sum=Net_sum[['df_networth']].sort_values("df_networth",ascending=False)
Net_sum.head(15)
```

```python
Net_sum.tail(15)
```

```python
fig,ax=plt.subplots(figsize=(16,8))
ax.vlines(x=Net_sum.index,ymin=0,ymax=Net_sum.df_networth,color='red',linewidth=5,alpha=0.6)
ax.scatter(x= Net_sum.index,y=Net_sum.df_networth,s=200,color='blue')
ax.set_xticks(Net_sum.index )
# ax.set_xticklabels(Net_sum.Country.str.upper(),rotation=90,fontsize=15)
# plt.scatter(Net_sum['df_networth'],df['Country'].index)
plt.show()
```
